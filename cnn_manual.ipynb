# CNN sur classification manuelle 

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
```

# code pour préparer X et y

```python
df = pd.read_csv("dataset_500Hz.csv")
df.head()
```

```python
# Récupérer les signaux
hpc = df["hpc_500"].values
pfc = df["pfc_500"].values
labels = df["manual_500"].values

# Paramètres
fs = 500
window_s = 5
window = fs * window_s  # 2500 points
step = window           # pas = 2500 (fenêtres non superposées)

X = []
y = []

for start in range(0, len(hpc) - window, step):
    end = start + window
    
    # Signal segmenté
    segment_hpc = hpc[start:end]
    segment_pfc = pfc[start:end]
    
    # Matrice : (2500, 2)
    segment = np.vstack([segment_hpc, segment_pfc]).T
    
    # Label = label central
    center_label = labels[start + window // 2]
    
    X.append(segment)
    y.append(center_label)

X = np.array(X)
y = np.array(y)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Classes:", np.unique(y))
```
