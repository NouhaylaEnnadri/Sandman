## Load one real file 

```python
from pathlib import Path
import scipy.io as sio

file_path = Path(r"R:\1\CN\post_trial1_2017-09-28_11-30-59\PFC_100_CH11.continuous.mat")

try:
    data = sio.loadmat(file_path, squeeze_me=True, simplify_cells=True)
    print("✅ File loaded successfully with scipy.io.loadmat")
    print("Top-level keys:", [k for k in data.keys() if not k.startswith('__')])
except NotImplementedError:
    import h5py
    with h5py.File(file_path, 'r') as f:
        print("✅ File loaded successfully with h5py (MAT v7.3 format)")
        print("Top-level groups:", list(f.keys()))
```
